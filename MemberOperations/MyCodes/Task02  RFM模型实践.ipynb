```python
import time 
import numpy as np
import pandas as pd
#引入机器学习包sklearn,sklearn==0.21.3
from sklearn.ensemble import RandomForestClassifier
```

```python
#创建sheet_names列表，与表格数据相对应，方便后续处理
sheet_names = ['2015','2016','2017','2018','会员等级']
#利用pandas库中的read_excel()函数，根据每一列的sheet_names,从第一行到最后一行依次读取每一行的数据作为sheet_datas
sheet_datas = [pd.read_excel('../Data/sales.xlsx',sheet_name=i) for i in sheet_names]#必须要保证与jupyter在同一个目录下才可以直接read_excel（'name.xlsx'）
```

```python
sheet_datas
```

```python
for i,data in enumerate(sheet_datas[:-1]):
    sheet_datas[i] = data.dropna(how='all',axis=0)
    sheet_datas[i] = data[data['订单金额'] > 1]# 丢弃订单金额<=1的记录
    sheet_datas[i]['max_year_date'] = data['提交日期'].max() # 增加一列最大日期值# 增加一列最大日期值
sheet_datas[0]
```

```python
sheet_datas
```

```python
for i in range(len(sheet_datas)):
    sheet_datas[i].dropna(how='any',axis=1,inplace=True)
sheet_datas[0]
```

```python
sheet_datas
```

```python
# 汇总所有数据
data_merge = pd.concat(sheet_datas[:-1],axis=0)
```
